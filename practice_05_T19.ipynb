```python
import numpy as np

def GSM (A,b, tol=0.000000001, max_iterations=100000):
    x=np.ones(len(b))
    for i in range (max_iterations):
        prev_x=np.copy(x)
        for j in range (len(b)):
            x[j]=(b[j]-np.dot(A[j,:],x)+x[j]*A[j,j])/A[j,j]
        if np.linalg.norm(x-prev_x)<tol:
            break
    
    return x
```

```python
# You can test your code here
A = np.array([[4.0, 1.0, 2.0],
              [3.0, 5.0, 1.0],
              [1.0, 1.0, 3.0]])
b = np.array([4.0, 7.0, 3.0])

x = GSM(A, b)
print("Solution (loop-based):", x)
x=np.linalg.solve(A,b)
print(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Get all colormap names
maps = plt.colormaps()

# Create gradient data for display
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))

# Plot them all
fig, axes = plt.subplots(nrows=len(maps)//4 + 1, ncols=4, figsize=(12, len(maps)//2))
axes = axes.flatten()

for i, name in enumerate(maps):
    ax = axes[i]
    ax.imshow(gradient, aspect='auto', cmap=plt.get_cmap(name))
    ax.set_title(name, fontsize=8)
    ax.axis('off')

# Hide unused subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
